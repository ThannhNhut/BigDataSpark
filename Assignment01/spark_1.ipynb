```
!pip install pyspark
import pyspark

from pyspark import SparkContext, SparkConf
import collections

conf = SparkConf().setMaster("local").setAppName("word couting")
sc = SparkContext.getOrCreate(conf=conf)

text = "to be or not be".split()
rdd = sc.parallelize(text)

counts = rdd.map(lambda word: (word,1))
print(counts.collect())

red = counts.reduceByKey(lambda x,y:x+y)
print(red.collect())

# min, max, average in array
numbers = [1, 4, 6, 2, 9, 10]
max_item = sc.parallelize(numbers).max()
min_item = sc.parallelize(numbers).min()
average = sc.parallelize(numbers).sum()/sc.parallelize(numbers).count()
print(max_item)
print(min_item)
print(average)



print(reduce(lambda x,y: "(" + str(x) + ", " + str(y) + ")",numbers))

f = lambda x,y: ["PASS",x,y] if x>3 and y<100 else ["FAIL",x,y]
print(f(4,50))
```
